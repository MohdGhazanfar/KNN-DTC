# Decision Tree Classifier

# DTC using entropy criterion VS DTC using the gini criterion.

```python
#imports
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import pandas as pd
```

Loading the iris dataset using scikit-learn.

```python
data = load_iris()
```

Splitting the dataset into training and testing sets with 70% of the data for training and 30% for testing.

```python

X = pd.DataFrame(data=data.data, columns=data.feature_names)
y = pd.Series(data=data.target, name='target')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) #With random_state=42 , we get the same train and test sets across different executions

```

```python
X.head()
```

Training a decision tree classifier using the entropy criterion and evaluate its accuracy on the testing set.

```python
DTC_MODEL_E = DecisionTreeClassifier(criterion='entropy', max_depth=6,random_state=0)
DTC_MODEL_E.fit(X_train, y_train)
accuracy_E = DTC_MODEL_E.score(X_test, y_test)

print(accuracy_E)
```

Training another decision tree classifier using the gini criterion and evaluate its accuracy on the testing set.

```python
DTC_MODEL_G = DecisionTreeClassifier(criterion='gini', max_depth=6, random_state=0)
DTC_MODEL_G.fit(X_train, y_train)
accuracy_G = DTC_MODEL_G.score(X_test, y_test)

print(accuracy_G)
```

Comparing the performance of the two classifiers and discuss the differences in terms of accuracy.

```python
if (accuracy_E> accuracy_G):
    print("Entropy criterion decision tree is more accurate than gini criterion.")
elif (accuracy_E< accuracy_G):
    print("Gini criterion decision tree is more accurate than entropy criterion.")

else:
     print("Both criterion decision tree has same accuracy.")
    
```

 Visualizing the decision tree for each classifier and comparing the tree structure and feature importance.

```python
plt.figure(figsize=(10,9))
plot_tree(DTC_MODEL_E, filled=True, feature_names=data.feature_names, fontsize=10)
print(DTC_MODEL_E.feature_importances_)
```

```python
plt.figure(figsize=(10,9))
plot_tree(DTC_MODEL_G, filled=True, feature_names=data.feature_names, fontsize=10)
print(DTC_MODEL_G.feature_importances_)

```
